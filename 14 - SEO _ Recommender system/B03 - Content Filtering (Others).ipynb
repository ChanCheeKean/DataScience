```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

# 1) Cosine Distance

```python
training_set = pd.read_csv(
    'ml-100k/u1.base', delimiter='\t', names=['user_id', 'movie_id', 'rating', 'time'])
matrix_train = training_set.groupby(['user_id', 'movie_id'])['rating'].mean().unstack().fillna(0)
matrix_array = matrix_train.to_numpy()
matrix_train.head()
```

```python
user_id = 0
result = cosine_similarity(np.expand_dims(matrix_array[user_id], axis=0), matrix_array)[0]
top_ids = np.argsort(result)[::-1][:10]
matrix_train.iloc[top_ids, :]
```
